```python
# Age
plt.figure(figsize=(15,10))
sns.boxplot('Pclass','Age',data=train_df)

# # Age
# plt.figure(figsize=(15,10))
# sns.boxplot('Pclass','Age',data=test_df)
```

```python
# fill nulls for Age
def Age_1(cols) :
    Age=cols[0]
    Pclass=cols[1] 
    
    if pd.isnull(Age) :
        if Pclass==1 :
            return 38
        elif Pclass==2:
            return 29
        else :
            return 24
    else :
        return Age


def Age_1(cols) :
    Age=cols[0]
    Pclass=cols[1] 
    
    if pd.isnull(Age) :
        if Pclass==1 :
            return 42 
        elif Pclass==2:
            return 25
        else :
            return 22
    else :
        return Age
    
    
test_df['Age']=test_df[['Age','Pclass']].apply(Age_1,axis=1)
train_df['Age']=train_df[['Age','Pclass']].apply(Age_1,axis=1)
```

```python
# remove rows with null 'Embarked'
train_df = train_df[pd.notnull(train_df['Embarked'])]
```

```python
# fill null fares based on Pclass
pclass_fare = test_df[['Pclass','Fare']].groupby('Pclass').median()
fare_dict = pclass_fare.to_dict()['Fare']
test_df['Fare'].fillna(fare_dict[test_df['Pclass'].iloc[0]], inplace=True)
```

```python
# create title feature
def name_to_title(dataset):
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['Title'] = dataset['Title'].replace(['Capt', 'Col', 'Major', 'Lady', 'Countess', 'Dr', 'Rev', 'Sir', 'Jonkheer'], 'Rare')
    dataset['Title'] = dataset['Title'].replace(['Ms','Mlle'], 'Miss')
    dataset['Title'] = dataset['Title'].replace(['Mme','Dona'], 'Mrs')
    dataset['Title'] = dataset['Title'].replace(['Don'], 'Mr')

name_to_title(train_df)
name_to_title(test_df)
```

```python
# create familysize feature
train_df['Familysize'] = train_df['SibSp'] + train_df['Parch'] + 1
test_df['Familysize'] = test_df['SibSp'] + test_df['Parch'] + 1
```

```python
# create new feature for fare bins
# def transform_fare(dataset):
#     dataset['Fare_bin'] = pd.cut(dataset['Fare'], bins=[0,7.8958,14.4542,31,513], 
#                                  labels=['fare_25pct','fare_50pct','fare_75pct','fare_100pct'])

def transform_fare(dataset):
    dataset['Fare_bin'] = pd.cut(dataset['Fare'], bins=[0,14.4542,31,513], 
                                 labels=['fare_low','fare_medium','fare_high'])

transform_fare(train_df)
transform_fare(test_df)
```

```python
# # create new feature for age bins
# train_df['Age_bin'] = pd.cut(train_df['Age'], bins=[0,18,40,80], 
#                                 labels=['Age_child','Age_adult','Age_seniors'])    

# test_df['Age_bin'] = pd.cut(test_df['Age'], bins=[0,18,40,80], 
#                                 labels=['Age_child','Age_adult','Age_seniors'])      
```

```python
# create bins for features

sex_bin = pd.get_dummies(train_df['Sex'],drop_first=True)
embarked_bin = pd.get_dummies(train_df['Embarked'])
fare_bin = pd.get_dummies(train_df['Fare_bin'])
title_bin = pd.get_dummies(train_df['Title'])
# age_bin = pd.get_dummies(train_df['Age_bin'])

# train_df = pd.concat([train_df, sex_bin, embarked_bin, fare_bin, title_bin, age_bin], axis=1)
train_df = pd.concat([train_df, sex_bin, embarked_bin, fare_bin, title_bin], axis=1)


sex_bin = pd.get_dummies(test_df['Sex'],drop_first=True)
embarked_bin = pd.get_dummies(test_df['Embarked'])
fare_bin = pd.get_dummies(test_df['Fare_bin'])
title_bin = pd.get_dummies(test_df['Title'])
# age_bin = pd.get_dummies(test_df['Age_bin'])
                      
# test_df = pd.concat([test_df, sex_bin, embarked_bin, fare_bin, title_bin, age_bin], axis=1)
test_df = pd.concat([test_df, sex_bin, embarked_bin, fare_bin, title_bin], axis=1)
```

```python
# Final table
# final_train = train_df.drop(['SibSp','Parch','PassengerId','Ticket','Cabin','Name','Sex','Ticket','Fare','Fare_bin','Title','Embarked','Age','Age_bin'], axis=1)
# final_test = test_df.drop(['SibSp','Parch','PassengerId','Ticket','Cabin','Name','Sex','Ticket','Fare','Fare_bin','Title','Embarked','Age','Age_bin'], axis=1)
final_train = train_df.drop(['SibSp','Parch','PassengerId','Ticket','Cabin','Name','Sex','Ticket','Fare','Fare_bin','Title','Embarked'], axis=1)
final_test = test_df.drop(['SibSp','Parch','PassengerId','Ticket','Cabin','Name','Sex','Ticket','Fare','Fare_bin','Title','Embarked'], axis=1)
```

# Model Prediction

```python
# split train_df further into subsets for training and testing
all_features = final_train.drop("Survived",axis=1)
target = final_train["Survived"]
X_train,X_test,y_train,y_test = train_test_split(all_features,target,test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=None, max_features='auto', max_leaf_nodes=None,
            min_samples_leaf=1,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            n_estimators=10000, n_jobs=-1, oob_score=False, random_state=0,
            verbose=0, warm_start=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(model.score(X_train, y_train))
print(accuracy_score(y_pred,y_test))
print(cross_val_score(model,all_features,target,cv=10,scoring='accuracy').mean())
```

```python
from sklearn.feature_selection import SelectFromModel
# features that have an importance of more than 0.15
sfm = SelectFromModel(model, threshold=0.05)

# Train the selector
sfm.fit(X_train, y_train)

# Print the name and gini importance of each feature
for feature in zip(X_train.columns,model.feature_importances_):
    print(feature)
    
X_important_train = sfm.transform(X_train)
X_important_test = sfm.transform(X_test)
# Create a new random forest classifier for the most important features
model_important = RandomForestClassifier(criterion='gini', n_estimators=400,
                             min_samples_split=3,min_samples_leaf=3,oob_score=True)

# Train the new classifier on the new dataset containing the most important features
model_important.fit(X_important_train, y_train)
y_pred = model_important.predict(X_important_test)
# print(model_important.score(X_train, y_train))
print(accuracy_score(y_pred,y_test))
# print(cross_val_score(model_important,all_features,target,cv=10,scoring='accuracy').mean())
```
