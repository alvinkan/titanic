# Kaggle Competition - Titanic Survival Prediction

# Wrangle dataset

```python
#drop cabin column
train_df.drop('Cabin', axis='columns', inplace=True)
test_df.drop('Cabin', axis='columns', inplace=True)
```

```python
# fill missing age data with median
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
test_df['Age'].fillna(test_df['Age'].median(), inplace=True)

# fill missing Embarked data with mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

# fill missing Fare data based on Pclass 
pclass_fare = test_df[['Pclass','Fare']].groupby('Pclass').median()
fare_dict = pclass_fare.to_dict()['Fare']
test_df['Fare'].fillna(fare_dict[test_df['Pclass'].iloc[0]], inplace=True)

# check nulls
print(train_df.isnull().sum())
print('-'*30)
print(test_df.isnull().sum())
```

```python
# combine dataset for data transformation
all_df = [train_df, test_df]
```

```python
# create new feature called Familysize
for dataset in all_df:
    dataset['Familysize'] = dataset['SibSp'] + dataset['Parch'] + 1
```

```python
# create new feature called title
for dataset in combine:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 
                                                 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
# create new feature for age bins
for dataset in all_df:
    dataset['Age_bin'] = pd.cut(dataset['Age'], bins=[0,12,20,25,30,40,65,80], 
                                labels=['Age_Child','Age_Teenager','Age_early20s','Age_late20s','Age_30s','Age_seniors','Age_retirees'])
# for dataset in all_df:
#     dataset['Age_bin'] = pd.cut(dataset['Age'], bins=[0,12,20,40,120], labels=['Children','Teenage','Adult','Elder'])
```

```python
# create new feature for fare bins
for dataset in all_df:
    dataset['Fare_bin'] = pd.cut(dataset['Fare'], bins=[0,7.91,14.45,31,513], labels=['Low_fare','median_fare',
                                                                                      'Average_fare','high_fare'])
```

```python
for dataset in all_df:
    drop_column = ['Age','Fare','Name','Ticket']
    dataset.drop(drop_column, axis=1, inplace = True)
```

```python
drop_column = ['PassengerId']
train_df.drop(drop_column, axis=1, inplace = True)
```

```python
train_df = pd.get_dummies(train_df, columns = ["Sex","Title","Age_bin","Embarked","Fare_bin"],
                             prefix=["Sex","Title","Age_type","Embarked_type","Fare_type"])
```

```python
test_df = pd.get_dummies(test_df, columns = ["Sex","Title","Age_bin","Embarked","Fare_bin"],
                             prefix=["Sex","Title","Age_type","Embarked_type","Fare_type"])
```

# Prediction model

```python
# split train_df further into subsets for training and testing
all_features = train_df.drop("Survived",axis=1)
target = train_df["Survived"]
X_train,X_test,y_train,y_test = train_test_split(all_features,target,test_size=0.3,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape

# initiate list to record scores of each model
models = []

# functiont to fit model, predict and record scores
def predict(model, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    train_score = model.score(X_train, y_train)
    pred_acc_score = accuracy_score(y_pred,y_test)
    cv_score = cross_val_score(model,all_features,target,cv=10,scoring='accuracy')

    models.append([model_name, train_score, pred_acc_score, cv_score.mean()])
    
    print('Training_dataset_score = ' + str(round(train_score * 100, 2)))
    print('Accuracy score = ' + str(round(pred_acc_score,4)))
    print('Cross validation score = ' + str(round(cv_score.mean(),4)))
```

### Logistic Regression

```python
#logistic regression
model = LogisticRegression()
predict(model, 'Logistic Regression')
```

### Random Forest

```python
model = RandomForestClassifier(criterion='gini',
            min_samples_leaf=1, min_samples_split=2,
            n_estimators=600)
predict(model, 'Random Forest')
```

### Support Vector Classification

```python
model = SVC()
predict(model, 'Support Vector Classification')
```

### KNN Classifier

```python
# finding optimal k
from sklearn.cluster import KMeans 
from scipy.spatial.distance import cdist 

K = range(1, 20) 
meandistortions = [] 
for k in K: 
    kmeans = KMeans(n_clusters=k) 
    kmeans.fit(X_train) 
    meandistortions.append(sum(np.min(cdist(X_train, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X_train.shape[0]) 
plt.plot(K, meandistortions, 'bx-') 
plt.xlabel('k') 
plt.ylabel('Average distortion') 
plt.title('Selecting k with the Elbow Method') 
plt.show() 

# knn model
model = KNeighborsClassifier(n_neighbors=9)
predict(model, 'KNN')
```

### Gaussian Naive Bayes

```python
model = GaussianNB()
predict(model, 'Gaussian Naive Bayes')
```

### Decision Tree

```python
model= DecisionTreeClassifier(criterion='gini', 
                               min_samples_split=10,min_samples_leaf=1,
                               max_features='auto')
predict(model, 'Decision Tree')
```

### Perceptron

```python
model = Perceptron()
predict(model, 'Perceptron')
```

### Stochastic Gradient Descent

```python
model = SGDClassifier()
predict(model, 'Stochastic Gradient Descent')
```

### Gradient Boosting

```python
model = GradientBoostingClassifier()
predict(model, 'Gradient Boosting Classifier')
```

### Model Evaluation

```python
models_df = pd.DataFrame(models,columns=['Model','Training_Acc_Score','Prediction_Score','Cross_Validated_Score'])
models_df.sort_values(by='Training_Acc_Score', ascending=False)
```

```python
# Prepare model dataset
X_train = train_df.drop('Survived', axis=1)
y_train = train_df['Survived']
X_test = test_df.drop('PassengerId', axis=1).copy()
X_train.shape, y_train.shape, X_test.shape
```

```python
# use random forest
model = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=None, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=400, n_jobs=1,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(model.score(X_train,y_train))

# prepare submission data
submission = pd.DataFrame({
        "PassengerId": test_df["PassengerId"],
        "Survived": y_pred
    })

# output to csv
submission.to_csv('submission.csv', index=False)
```
